## Bag of Words

```python
# cleaning the given input text before getting bag of words

import re
# importing all stemmer, lemmatizer, stopwords
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
ps = PorterStemmer()
wordnet = WordNetLemmatizer()
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
    
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
x = cv.fit_transform(corpus).toarray()
```

# Word2Vec

```python
from gensim.models import Word2Vec
```

```python
# preprocessing the paragraph just in case to remove unnecessary 
# format of txt like numbers, speacial chars, chars, and some uppercase
# and lower case text to convert it to lower case. to get rid of duplicates

text = re.sub(r'\[[0-9]\]', ' ', paragraph)
text = re.sub(r'\s+', ' ', text)
text = text.lower()
text = re.sub(r'\d', ' ', text)
text = re.sub(r'\s+', ' ', text)
```

```python
# tokenising the paragraph to sentecnes

sentences = nltk.sent_tokenize(text)
len(sentences)
```

```python
sentences
```

```python
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sentences
```

```python
# stopwords are removed

for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
    
    
sentences    
```

```python
model = Word2Vec(sentences, min_count = 1)
```

```python
words = model.wv
words
```
